###Toliau naudosiu failą iš internetinio adreso

```
import pandas as pd
#data = pd.read_csv('/content/drive/My Drive/Colab Notebooks/DAM/Fremont_Bridge_Hourly_Bicycle_Counts_by_Month_October_2012_to_present.csv',
#                  index_col='Date', parse_dates=True)
data = pd.read_csv('Fremont.csv', index_col='Date', parse_dates=True)
data.head()
```

```
%matplotlib inline
data.plot()
```

Susumuosim po kiek pravažiuoja per savaitę

```
data.resample('W').sum().plot()
```

###Panašu, kad *Google Colaboratory* jau iš karto yra nustačiusi kažką panašaus į „seaborn“ stilių

```
import matplotlib.pyplot as plt
plt.style.use('seaborn')
data.resample('W').sum().plot()
```

Sutruminam legendos informaciją

```
data.columns = ['West', 'East']
data.resample('W').sum().plot()
```

Norim pažiūrėti, ar yra kažkokių ilgalaikių tendencijų (per metus).

Kiekvienas grafiko taškas bus suma dviračių, pravažiavusių per praeitas 365 dienas.

```
ax = data.resample('D').sum().rolling(365).sum().plot()
ax.set_ylim(0, None)
```

Pridedam bendrą skaičių (susumuota per abi puses)

```
data['Total'] = data['West'] + data['East']
ax = data.resample('D').sum().rolling(365).sum().plot()
ax.set_ylim(0, None)
```

Skaičiuojam vidurkius kiekvienam paros laikui

```
data.groupby(data.index.time).mean().plot()
```

Pertvarkom duomenis taip, kad rodytų srautus kiekvieną valandą padieniui

```
pivoted = data.pivot_table('Total', index=data.index.time, columns=data.index.date)
pivoted.iloc[:5, :5]
```

Išbrėžiam kiekvienos dienos dviračių srautus skirtingomis valandomis

```
pivoted.plot(legend=False, alpha=0.01)
```
